```python
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
ratings_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
```

```python
#Reading CSV Files
import numpy as np
import pandas as pd
basics = pd.read_csv(basics_url,sep='\t', low_memory=False)
akas = pd.read_csv(akas_url, sep = '\t', low_memory=False)
ratings = pd.read_csv(ratings_url, sep = '\t', low_memory =False)
```

```python
#Replacing null cells
ratings = ratings.replace({'\\N':np.nan})
akas = akas.replace({'\\N':np.nan})
basics = basics.replace({'\\N':np.nan})
```

```python
basics1 = basics1[basics1['startYear'].notna()]
basics1['startYear'].isnull().sum()
```

```python
basics1['startYear'] = basics1['startYear'].astype(str).astype(int)
```

```python
basics1.head()
```

```python
basics2 = basics1[basics1['startYear']>=2000]
basics2 = basics2[basics2['startYear']<2022]
basics2.head()
```

```python
akas1 = akas[akas['region']=='US']
akas1.head()
```

```python
is_documentary = basics2['genres'].str.contains('documentary',case=False)
basics2 = basics2[~is_documentary]
```

```python
keepers =basics2['tconst'].isin(akas1['titleId'])
keepers
```

```python
basics2 = basics2[keepers]
basics2
```
